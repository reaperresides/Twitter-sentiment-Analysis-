# Data preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk

nltk.download("stopwords")
```

```python
cols = ("label",	"UserID"	,"Date"	,"Query"	,"Username"	,"Tweet")

data = pd.read_csv("/content/drive/MyDrive/training.1600000.processed.noemoticon.csv",encoding='latin-1',names=cols)
data.head()
```

```python
data.label.value_counts()
```

```python
def Randomize(data):
  indexes = len(data)
  new_index = np.random.choice(indexes,size=len(data))
  data = data.iloc[new_index,:]
  return data



data =Randomize(data)
data.head()
```

```python
features_data = data.iloc[:,5]
features_data.head()
```

```python
import re

final_data = []
for message in features_data:
  b = re.sub("(@\w+)"," ",message)  #excluding tagged data
  c = re.sub("(http)\D+\w+"," ",b)  #excluding hyperlinks
  d = re.sub("[^A-Za-z]"," ",c)     #excluding white space characters
  e = d.split(" ")                  #making a list of data
  final_data.append([i for i in e if i!=''])
  
```

```python
final_data
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords


ps = PorterStemmer()

corpus = []



#taking 100k data points
for word in final_data[:100000]:
  row = [ps.stem(w) for w in word if w not in stopwords.words("english")]
  corpus.append(row)

```

```python
final_data_new = [" ".join(x) for x in corpus]
final_data_new
```

```python
Y = data.label[:100000].values
Y = np.where(Y==4,1,0)
print(np.unique(Y,return_counts=True))
```

```python
#creating series of lemmatized data

data_pd = pd.Series(data=final_data_new,name="data")   

#creating series of mapped labels
label_pd = pd.Series(data=Y.ravel(),name="labels")

#creating dataframe of new data
new_df = pd.concat([data_pd,label_pd],axis=1)

new_df


#selecting only those sentences having length greater than 5
filt = new_df["data"].map(lambda x:len(x.split(" ")) > 5)
final_df = new_df.loc[filt]
final_df


X_data = final_df["data"].values
Y_data = final_df["labels"].values
```
